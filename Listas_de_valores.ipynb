##Lista

>Uma das principais características de uma lista é que os valores podem ser modificados

```python
lista = [7, 'Gustavo', 2.5]
```

```python
type(lista)
```

```python
lista.append('Duarte') # Adiciona um elemento no último lugar da lista
```

```python
print(lista)
print(lista[1])
```

##Tupla

>Uma das principais diferenças entre tuplas e listas é que os valores das tuplas não podem ser modificados

```python
tupla = (1, 'Gustavo', 9.3)
```

```python
type(tupla)
```

```python
print(tupla)
```

##Dicionário

>dicionários são quase como listas só que ao invés de conterem indices ele contêm chaves e campos

```python
dicionario = {'nome': 'Gustavo', 'idade': 22}
```

```python
type(dicionario)
```

```python
print(dicionario)
```

```python
print(dicionario['nome'])
```

```python
dicionario['sobrenome'] = 'Duarte'
```

```python
print(dicionario)
```
