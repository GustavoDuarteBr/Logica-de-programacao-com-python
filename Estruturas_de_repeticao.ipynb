##for

```python
# for range
for contador in range(10):
  print(contador)
```

```python
# for range + break
for contador in range(10000):
  print(contador)
  if contador == 2:
    break # break é utilizado para quebrar o loop
```

```python
lista = ['teste1', 'teste2']
```

```python
# for simples
for val in lista:
  print(val)
```

```python
# for com enumerate
for i, val in enumerate(lista):
  print(i, val) # enumerate é utilizado para descobrir o índice dos valores
```

##for com operadores compostos

```python
lista2 = [1,7,21,59]
```

```python
for val in lista2:
  val += 3
  print(val)
```

```python
for val in lista2:
  val *= 9
  print(val)
```

##while

```python
val = 1
while val <= 9:
  val += 2
  print(val)
```

```python
#while com else
val = 1
while val < 73:
  val *= 4
else:
  print(f'O valor de val é: {val}')
```

```python
#while com break
val = 1
while True:
  print(val)
  val += 1
  if val > 8:
    break
```
