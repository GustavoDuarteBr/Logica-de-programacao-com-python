Type

* int - São os números inteiros. ex: 1, 2, 3

* float - São os números com casas decimais. ex: 1.23, 12.00, 34.1

* str (string) - São uma sequência de caracteres (letras, números ou símbolos) ex: 'Olá', '1234', 'Meu nome é Gustavo'

* bool (boolean) - Representa valores lógicos, usados para decisões e condições.
Possui apenas dois valores possíveis: true (verdadeiro) ou false (falso) ex: 2 == 2 vai retornar true, 5 != 5 vai retornar false

```python
x = 10
type(x)
print(f'{x} {type(x)}')

nome = 'Gustavo'
print(f'{nome} {type(nome)}')

num = '100'
num_convert = int(num) # Converte a variável num que é uma string para um int
num_float = float(num) # Converte a variável num que é uma string para um float
print(f'{num} {type(num)}')
print(f'{num_convert} {type(num_convert)}')
print(f'{num_float} {type(num_float)}')
```

```python
valor = input('Digite o valor do produto: ')

valor_convert = float(valor)

desconto = valor_convert - (valor_convert * (5/100)) # utiliza-se parênteses para enfatizar a conta que você quiser

print(f'O valor de {valor} irá receber um desconto de 5% e ficará custando {desconto}')
```
