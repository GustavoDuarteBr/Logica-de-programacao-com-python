#Condicional

```python
# if
if 5 < 10:
  print('Verdade')
```

```python
# if/else
num1 = 10
num2 = 31

if num1 > num2:
  print(f'{num1} é maior que {num2}')
else:
  print(f'{num2} é maior que {num1}')
```

```python
# if/elif/else
num1 = 4
num2 = 7

if num1 > num2:
  print(f'{num1} é maior que {num2}')
elif num2 > num1:
  print(f'{num2} é maior que {num1}')
else:
  print(f'{num1} e {num2} são iguais')
```

#Condicional com operadores lógicos

```python
num1 = 39
num2 = 14
num3 = 37

if num1 > num2 and num1 > num3:
  print(f'{num1} é o maior número')
else:
  print(f'{num1} não é o maior número')
```
