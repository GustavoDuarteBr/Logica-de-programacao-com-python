```python
while True: # com while true cria um loop infinito
  primeiro_numero = float(input('Digite o primeiro número: ')) # usa o input para o usuário digitar o primeiro número
  operacao = int(input('1-soma \n2-subtração \n3-multiplicação \n4-divisão \n5-divisão inteira \n6-módulo \nEscolha a operação desejada: ')) # usa o input para o usuário digitar a operação desejada
  segundo_numero = float(input('Digite o segundo número: ')) # usa o input para o usuário digitar o segundo número

  if operacao == 1: # vê a operação selecionada e realiza a soma
    resultado_soma = primeiro_numero + segundo_numero # soma dos dois número selecionados
    print(f'{primeiro_numero} + {segundo_numero} = {resultado_soma}')

  elif operacao == 2: # vê a operação selecionada e realiza a subtração
    resultado_sub = primeiro_numero - segundo_numero
    print(f'{primeiro_numero} - {segundo_numero} = {resultado_sub}')

  elif operacao == 3: # vê a operação selecionada e realiza a multiplicação
    resultado_mult = primeiro_numero * segundo_numero
    print(f'{primeiro_numero} * {segundo_numero} = {resultado_mult}')

  elif operacao == 4: # vê a operação selecionada e realiza a divisão
    resultado_divi = primeiro_numero / segundo_numero
    print(f'{primeiro_numero} / {segundo_numero} = {resultado_divi}')

  elif operacao == 5: # vê a operação selecionada e realiza a divisão inteira
    resultado_divi_inteira = primeiro_numero // segundo_numero
    print(f'{primeiro_numero} // {segundo_numero} = {resultado_divi_inteira}')

  elif operacao == 6: # vê a operação selecionada e realiza o módulo
    resultado_modul = primeiro_numero % segundo_numero
    print(f'{primeiro_numero} % {segundo_numero} = {resultado_modul}')

  else: # vê a operação selecionada e diz se é inválida
    print(f'{operacao} é uma operação inválida')

  continuar = input('Deseja fazer outra conta? \n[S] ou [N]: ') # pergunta ao usuário se ele quer quebrar o loop

  if continuar.upper() == 'N': # quebra o loop
    break
```
