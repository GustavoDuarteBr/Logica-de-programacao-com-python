## Tabela Verdade

and

```python
print(10 > 2 and 3 != 4) # V and V = V
print(3 == 3 and 10 > 11) # V and F = F
print(5 != 5 and 1 == 4) # F and F = F
print(12 <= 10 and 9 >= 8) # F and V = F
```

or

```python
print(10 > 2 or 3 != 4) # V or V = V
print(3 == 3 or 10 > 11) # V or F = V
print(5 != 5 or 1 == 4) # F or F = F
print(12 <= 10 or 9 >= 8) # F or V = V
```

not

```python
not 10 < 2 # F se torna V
```

```python
not 10 > 2 # V se torna F
```

# Variáveis

Variáveis estão alocadas em espaços na memória

```python
# Nome da variável é "x" que armazena o valor 10 em um espaço na memória
x = 7

print(x) # Utilizar a referência no nome atribuido
```
